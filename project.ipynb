```python
#Scenario
#Imagine the following: you're walking up the empire state building 
# and you're playing a game with a friend. 
#You throw a die one hundred times.
# If it's 1 or 2 you'll go one step down.
# If it's 3, 4, or 5, you'll go one step up.
# If you throw a 6, you'll throw the die again and walk up the resulting 
#  number of steps.
#You admit that you're a bit clumsy and have a chance of 0.1% of falling down 
# the stairs when you make a move. Falling down means that you have to start 
# again from step 0. 
#With all of this in mind, you bet your friend that you'll reach 60 steps high.

#Question
#What is the chance that you will win this bet?
```

```python
import numpy as np
np.random.seed(123)

#Simulate a dice roll
print(np.random.randint(1, 7))
```

```python
#Constructing the if statement

np.random.seed(123)

step = 50

dice = np.random.randint(1, 7)

if dice <= 2 :
    step = step - 1
elif dice > 2 and dice < 6 :
    step = step + 1
else :
    step = step + np.random.randint(1,7)

print(dice) , print(step)
```

```python
#Generating a Random Walk

np.random.seed(123)

random_walk = [0]

for x in range(100) :
    step = random_walk[-1]
    dice = np.random.randint(1,7)
     
    if dice <= 2:
        step = step - 1
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)
    random_walk.append(step)
print(random_walk)
```

```python
#Fixing ability for step to go below 0

np.random.seed(123)

random_walk = [0]

for x in range(100) :
    step = random_walk[-1]
    dice = np.random.randint(1,7)
     
    if dice <= 2:
        step = max(0, step - 1)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)
    random_walk.append(step)
print(random_walk)

#Visualize the random walk
import matplotlib.pyplot as plt

plt.plot(random_walk)
plt.show()
```

```python
#Simulating 10 times

np.random.seed(123)

all_walks = []

for i in range(10) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)

    all_walks.append(random_walk)
print(all_walks)

np_aw = np.array(all_walks)
np_aw_t = np.transpose(np_aw)

#Visualize All Walks

plt.plot(np_aw_t)
plt.show()
```

```python
#Simulating 250 times

np.random.seed(123)

all_walks = []
for i in range(250) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # Implement clumsiness
        if np.random.rand() <= .001 :
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)

#Visualize
np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.show()
```

```python
#Simulating 500 times

np.random.seed(123)

all_walks = []
for i in range(500) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001 :
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)

np_aw_t = np.transpose(np.array(all_walks))
#Getting the final step of each simulation
ends = np_aw_t[-1,:]
#Visualizing the last step
plt.hist(ends)
plt.show()
```

```python
#Answering the Question of the % Chance to win the bet

percentchance = np.mean(ends>59)
print(percentchance)
```
